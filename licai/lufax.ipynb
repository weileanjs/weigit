```python
import pandas as pd
import numpy as np
import time
import pymongo
import charts
client = pymongo.MongoClient('localhost',27017)
db = client['lufax']
```

```python
date='2016-12-12'
```

```python
data_all =pd.DataFrame(list(db['lu_zhuanr'].find({})))
data_all.pop('_id')
data = pd.DataFrame(list(db['lu_zhuanr'].find({'date':date})))
data.pop('_id')
data
```

```python
date_l = list(set(data_all['date']))
date_l.sort()
date_l
```

```python
data = pd.DataFrame(list(db['lu_zhuanr'].find({'date':date})))
```

```python
a = []
b = []
c = []
d = []
for x in date_l:
    data = pd.DataFrame(list(db['lu_zhuanr'].find({'date':x})))
    bins = [0, 120000, 220000, 330000, 1000000]
    group_price = ['0~12万', '12~22万', '22~33万', '33万+']
    data['price_group'] = pd.cut(data['price'].astype(float), bins, labels=group_price)
    data['interest']=data['interest'].astype(float)
    data_g_inser_mean = data.groupby('price_group')['interest'].agg(np.mean)
    inser_index = list(data_g_inser_mean.index)
    inser_values = list(data_g_inser_mean.values)
    a.append(inser_values[0])
    b.append(inser_values[1])
    c.append(inser_values[2])
    d.append(inser_values[3])
print(a,b,c,d)
```

```python
series = [{ 'name': '1~12万','data': a}, {'name': '12~22万','data': b}, { 'name': '22~33万','data': c},{ 'name': '33万+','data': d}]
options = {
    'title'   : {'text': '陆金所财富汇转让区利率统计'},
    'subtitle': {'text': 'powered by W.C.H'},
    'xAxis': [{'categories': date_l,'crosshair': 'true'}],
    'yAxis': {'title': {'text': '利率 (%)'},
            'plotLines': [{'value': 0, 'width': 1,'color': '#808080'}]},

    }

charts.plot(series, show='inline',options=options)
```
